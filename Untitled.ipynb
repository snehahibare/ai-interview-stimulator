````python
import os
import json
import anthropic
import ipywidgets as widgets
from IPython.display import display, clear_output
from dotenv import load_dotenv

load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))
print("✅ Connected to Claude!")
````

````python
QUESTION_BANK = {
    "Software Engineer": [
        "Tell me about yourself.",
        "Explain REST vs GraphQL.",
        "How do you handle disagreement with your tech lead?",
        "Describe a bug you fixed.",
        "How do you write clean code?"
    ],
    "Data Scientist": [
        "Walk me through your data science workflow.",
        "How do you handle missing data?",
        "Explain overfitting and how to prevent it.",
        "Supervised vs unsupervised learning?",
        "How do you explain a model to non-technical people?"
    ],
    "Product Manager": [
        "How do you prioritize features?",
        "Tell me about a product you launched.",
        "How do you handle conflicting feedback?",
        "What metrics measure feature success?",
        "Describe writing a PRD."
    ],
    "General": [
        "Tell me about a time you failed.",
        "Describe a conflict with a coworker.",
        "What are your strengths and weaknesses?",
        "Where do you see yourself in 5 years?",
        "Why do you want to work here?"
    ]
}
print("Question bank loaded!")
````

````python
def get_ai_feedback(role, question, answer):
    prompt = f"""
You are an expert interviewer for the role of {role}.

The candidate was asked: "{question}"
Their answer was: "{answer}"

Respond ONLY in this JSON format:
{{
  "score": <integer 1-10>,
  "strengths": "<what they did well>",
  "weaknesses": "<what was missing>",
  "ideal_answer": "<a brief model answer>",
  "follow_up": "<one follow-up question>"
}}
"""
    response = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=800,
        messages=[{"role": "user", "content": prompt}]
    )
    raw = response.content[0].text
    clean = raw.strip().removeprefix("```json").removesuffix("```").strip()
    return json.loads(clean)

print("✅ Feedback engine ready!")
````

````python
session = {
    "role": None,
    "questions": [],
    "answers": [],
    "feedbacks": [],
    "current_index": 0
}

def start_session(role):
    session["role"] = role
    session["questions"] = QUESTION_BANK.get(role, QUESTION_BANK["General"])
    session["answers"] = []
    session["feedbacks"] = []
    session["current_index"] = 0

def get_current_question():
    idx = session["current_index"]
    if idx < len(session["questions"]):
        return session["questions"][idx]
    return None

def submit_answer(answer):
    q = get_current_question()
    feedback = get_ai_feedback(session["role"], q, answer)
    session["answers"].append(answer)
    session["feedbacks"].append(feedback)
    session["current_index"] += 1
    return feedback

print("✅ Session tracker ready!")
````

````python
role_dropdown = widgets.Dropdown(options=list(QUESTION_BANK.keys()), description="Role:")
start_btn = widgets.Button(description="Start Interview", button_style="success")
answer_box = widgets.Textarea(placeholder="Type your answer here...", layout=widgets.Layout(width="600px", height="120px"))
submit_btn = widgets.Button(description="Submit Answer", button_style="primary", disabled=True)
next_btn = widgets.Button(description="Next Question", button_style="warning", disabled=True)
end_btn = widgets.Button(description="End & See Report", button_style="danger", disabled=True)
output_area = widgets.Output()

def on_start(b):
    role = role_dropdown.value
    start_session(role)
    submit_btn.disabled = False
    end_btn.disabled = False
    clear_output(wait=True)
    with output_area:
        clear_output()
        q = session["questions"][0]
        print(f"🎙️ Interview started for: {role}")
        print(f"\n❓ Question 1: {q}\n")

start_btn.on_click(on_start)
print("✅ Fixed!")

def on_submit(b):
    answer = answer_box.value.strip()
    if not answer:
        with output_area:
            print("⚠️ Please write an answer first.")
        return
    with output_area:
        print("\n⏳ Analyzing your answer...")
    feedback = submit_answer(answer)
    answer_box.value = ""
    with output_area:
        clear_output()
        idx = session["current_index"]
        print(f"✅ Feedback for Q{idx}:")
        print(f"  🏅 Score      : {feedback['score']} / 10")
        print(f"  💪 Strengths  : {feedback['strengths']}")
        print(f"  🔧 Weaknesses : {feedback['weaknesses']}")
        print(f"  📖 Ideal Ans  : {feedback['ideal_answer']}")
        print(f"  🔄 Follow-up  : {feedback['follow_up']}")
        if get_current_question():
            next_btn.disabled = False
        else:
            print("\n🎉 All questions done! Click End & See Report.")

def on_next(b):
    next_btn.disabled = True
    with output_area:
        clear_output()
        idx = session["current_index"] + 1
        print(f"\n❓ Question {idx}: {get_current_question()}\n")

def on_end(b):
    scores = [f["score"] for f in session["feedbacks"]]
    avg = sum(scores) / len(scores) if scores else 0
    with output_area:
        clear_output()
        print("=" * 50)
        print("📊  INTERVIEW REPORT")
        print("=" * 50)
        print(f"Role      : {session['role']}")
        print(f"Questions : {len(session['answers'])}")
        print(f"Avg Score : {avg:.1f} / 10")
        print("-" * 50)
        for i, (q, a, f) in enumerate(zip(session["questions"], session["answers"], session["feedbacks"])):
            print(f"\nQ{i+1}: {q}")
            print(f"Your Answer : {a[:80]}...")
            print(f"Score       : {f['score']}/10")

start_btn.on_click(on_start)
submit_btn.on_click(on_submit)
next_btn.on_click(on_next)
end_btn.on_click(on_end)

display(widgets.VBox([
    widgets.HBox([role_dropdown, start_btn]),
    answer_box,
    widgets.HBox([submit_btn, next_btn, end_btn]),
    output_area
]))
````
